```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path("../data/raw")
raw_files = sorted(RAW_DIR.glob("*.csv"))
raw_files
```

```python
dfs = {
    f.stem: pd.read_csv(f, parse_dates=["Date"], index_col="Date")
    for f in raw_files
}
list(dfs.keys())
```

```python
# quick look per field file: shape, dtypes, missing values, date range, head()
for name, df in dfs.items():
    print(f"=== {name} ===")
    print("shape:", df.shape)
    print("securities:", list(df.columns))
    print("date range:", df.index.min(), "->", df.index.max())
    print("\ndtypes:")
    print(df.dtypes)
    print("\nmissing values per security:")
    print(df.isna().sum())
    print("\nhead():")
    display(df.head())
    print("\n")
```

```python
# per-security quick look, combining all fields (open/high/low/close/volume) into one frame
securities = dfs["close"].columns

for sec in securities:
    print(f"=== {sec} ===")
    sec_df = pd.DataFrame({name: df[sec] for name, df in dfs.items()})
    print("shape:", sec_df.shape)
    print("dtypes:")
    print(sec_df.dtypes)
    print("date range:", sec_df.index.min(), "->", sec_df.index.max())
    print("missing values:")
    print(sec_df.isna().sum())
    print("head():")
    display(sec_df.head())
    print("\n")
```

## Visual overview

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# 1. Data coverage: first/last available date per security (catches gaps or mismatched start dates)
fig, ax = plt.subplots(figsize=(9, 4))
close = dfs["close"]
for i, sec in enumerate(securities):
    s = close[sec].dropna()
    ax.plot([s.index.min(), s.index.max()], [i, i], marker="|", markersize=12, linewidth=4)
ax.set_yticks(range(len(securities)))
ax.set_yticklabels(securities)
ax.set_title("Data coverage by security")
ax.set_xlabel("Date")
fig.tight_layout()
```

```python
# 2. Normalized close price (rebased to 100) per pair - as defined in README.md
normalized = close / close.iloc[0] * 100
pairs = [("XOM", "CVX"), ("XLE", "XOP"), ("NEM", "GOLD"), ("ADM", "BG")]

fig, axes = plt.subplots(2, 2, figsize=(11, 8))
for ax, (a, b) in zip(axes.flat, pairs):
    normalized[[a, b]].plot(ax=ax)
    ax.set_title(f"{a} vs {b}")
    ax.set_xlabel("Date")
    ax.set_ylabel("Rebased price")
    ax.legend(loc="upper left", fontsize=8)
fig.suptitle("Close price, rebased to 100 at start")
fig.tight_layout()
```

```python
# 3. Daily returns distribution per security - sanity check for outliers / data errors and relative volatility
daily_returns = close.pct_change().dropna()
fig, ax = plt.subplots(figsize=(9, 5))
daily_returns.boxplot(ax=ax)
ax.set_title("Daily returns distribution")
ax.set_ylabel("Daily return")
fig.tight_layout()
```
